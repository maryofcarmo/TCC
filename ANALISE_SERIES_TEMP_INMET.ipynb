# 2° Carrego as séries temporais de cada usina meteorológica

```python
# Defina o diretório onde estão os arquivos CSV
diretorio = '/content/drive/MyDrive/TCC/DADOS/SERIES_TEMP_LIMPAS'

# Obtenha a lista de todos os arquivos CSV no diretório
arquivos_csv = [f for f in os.listdir(diretorio) if f.endswith('.csv')]

arquivos_csv
```

# 3° armazenos os 116 arquivos de séires temporais em um dicionário

## onde key: nome do  arquivo

```python
# dicionário para armazenar as séries temporais
series_temp = {}


# Loop para ler cada arquivo CSV e armazenar no dicionário
for arquivo in arquivos_csv:

    caminho_completo = os.path.join(diretorio, arquivo)

    #obtem o nome do arquivo sem a sua extensão e coloca na variável 'nome_df'
    nome_df = os.path.splitext(arquivo)[0]

    # Lê o arquivo CSV e armazena no dicionário com o nome do arquivo como chave
    series_temp[nome_df] = pd.read_csv(caminho_completo)

    # Setar a primeira coluna, sem dizer o nome da coluna, como o índice
    series_temp[nome_df].set_index(series_temp[nome_df].columns[0], inplace=True)

    # Converter o índice para datetime
    series_temp[nome_df].index = pd.to_datetime(series_temp[nome_df].index)
```

# 3° Retirar séries temporais inteiramente vazias do dicionário

```python
# Inicializar uma lista para armazenar os nomes das séries que estão vazias
series_vazias = []

# Iterar sobre cada série no dicionário 'series_temp'
for nome, df in series_temp.items():
    # Verificar se o DataFrame está vazio
    if df.empty:
        # Se estiver vazio, adicionar o nome da série à lista
        series_vazias.append(nome)

# Exibir a lista de séries vazias
for series in series_vazias:
    print(series)
```

```python
# quantidade de séries vazias
len(series_vazias)

# Exibindo o resultado
print(f"A quantidade de séries temporais vazias é: {len(series_vazias)}")
```

```python
# Removendo DE FATO as séries temporais vazias do dicionário
for key in series_vazias:
    series_temp.pop(key, None)

# Obtendo a quantidade de itens no dicionário
quantidade_itens = len(series_temp)

# Exibindo o resultado
print(f"A quantidade de séries temporais no dicionário agora é: {quantidade_itens}")
```

## CARREGO o arquivo CSV com as variáveis de correlação de cada usina

```python
# Defina o diretório onde estão os arquivos CSV
diretorio = '/content/drive/MyDrive/TCC/DADOS/INFORMACOES_ESTACOES/est_com_correlacao.csv'
# Obtenha a lista de todos os arquivos CSV no diretório

# Ler o arquivo CSV
info_completa_inmet = pd.read_csv(diretorio)

info_completa_inmet
```

## retiro as 23 usinas com valores vazios

```python
# Filtrar o DataFrame para remover as linhas presentes na lista de séries vazias
df_inmet_filtrados = info_completa_inmet[~info_completa_inmet['nome do arquivo'].isin(series_vazias)]

# Exibir o DataFrame filtrado
df_inmet_filtrados
```

## Encontrando a estação meteorológica com maior complementariedade solar-eólica (através do coeficiente de Pearson)

```python
# Encontrar o índice do valor mais próximo de -1 na coluna 'P3'
indice_mais_proximo_de_menos_um = (df_inmet_filtrados['P3'] - (-1)).abs().idxmin()

# Obter o valor correspondente de 'P3' e 'nome do arquivo'
valor_mais_proximo_de_menos_um = df_inmet_filtrados.loc[indice_mais_proximo_de_menos_um, 'P3']
nome_arquivo_correspondente = df_inmet_filtrados.loc[indice_mais_proximo_de_menos_um, 'nome do arquivo']

# Exibir os resultados
print(f"O valor mais próximo de -1 na coluna 'P3' é: {valor_mais_proximo_de_menos_um}")
print(f"Esse valor está associado ao arquivo: {nome_arquivo_correspondente}")

```
